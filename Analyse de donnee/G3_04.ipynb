```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('amazon.csv')
```

Dans la cellule ci dessous :

```python
print("Top 5 des meilleures catégories:")
top_categories = data['listed_in'].value_counts().head(5)
print(top_categories)
top_categories.plot(kind='bar')
plt.title('Top 5 des meilleures catégories')
plt.show()
```

Dans la cellule ci dessous :

recent_data = data[data['release_year'] > 2015] : 

    Cette ligne sélectionne les lignes de l’ensemble de données où l’année de sortie est supérieure à 2015. 
    Cela permet de se concentrer sur les données récentes.

tv_vs_movies = recent_data['type'].value_counts() : 

    Cette ligne compte combien de fois chaque type unique (par exemple, ‘TV Show’ ou ‘Movie’) apparaît dans les données récentes. 
    Cela donne une idée de la quantité de chaque type de contenu.

tv_vs_movies.plot(kind='bar') : 

    Cette ligne crée un graphique à barres des résultats. 

plt.title('Amazon se concentre-t-il davantage sur les séries télévisées que sur les films ces dernières années ?') : 

    Cette ligne ajoute le titre au graphique.

plt.show() :

    Enfin, cette ligne affiche le graphique à l’écran.


```python
print("Amazon se concentre-t-il davantage sur les séries télévisées que sur les films ces dernières années?")
recent_data = data[data['release_year'] > 2015]
tv_vs_movies = recent_data['type'].value_counts()
print(tv_vs_movies)
tv_vs_movies.plot(kind='bar')
plt.title('Amazon se concentre-t-il davantage sur les séries télévisées que sur les films ces dernières années ?')
plt.show()
```
